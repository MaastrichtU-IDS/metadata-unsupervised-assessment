```python
data = pd.read_csv('NCBItrainset_corpus1.tsv', sep='\t', header=None)
```

```python
data.columns = ['pubmed_id','value','class','id_ontology']
```

```python
data = data[~data['pubmed_id'].isnull()]
```

```python
table = data[~data['value'].isnull()]
```

```python
table.groupby('class').count()['pubmed_id']
```

```python
table = table[table['class'] != 'CompositeMention']
```

```python
vec_raw = data[data['value'].isnull()]['pubmed_id']
vec_split = []
for line in vec_raw:
    vec_split.append(line.split("|"))
```

```python
corpus = pd.DataFrame(vec_split)
corpus.columns = ['pubmed_id','type','text']
```

```python
df_titles = corpus[corpus['type'] == 't'][['pubmed_id','text']]
df_abstracts = corpus[corpus['type'] == 'a'][['pubmed_id','text']]
```
